```python
import pandas as pd
import matplotlib.pyplot as plt
# sample data
data = {
    'TIME':['08:30','09:15','12:30','1:45','7:00','08:15','9:00','12:00','6:30','8:45'],
    'LOCATION':['Bangalore','Madurai','Hyderabad','Madurai','Bangalore','Hyderabad','Madurai','Bangalore','Madurai','Hyderabad'],
    'ITEM':['Briyani','chicken rice','Burger','chicken rice','Briyani','Burger','chicken rice','Briyani','Burger','Briyani']
}
# create dataframe
df=pd.DataFrame(data)
print("--------------Food Delivery Data-------------------")
print(df)
#To Analayze order data
df['Hour'] = pd.to_datetime(df['TIME'], format='%H:%M').dt.hour
#To Analayze the peak orders
peak_times=df['Hour'].value_counts().sort_index()
print("\n ********Peak Order Times**********")
print(peak_times)
#To analyze the most popular dishes
Popular_dishes=df['ITEM'].value_counts()
print("\n********** Most Popular Dishes**********")
print(Popular_dishes.to_string())
#for order by location
location_orders=df['LOCATION'].value_counts()
print("\n ***********Orders by Location**********")
print(location_orders.to_string())
#for dashboard
plt.figure(figsize=(15,5))
#for peak order times
plt.subplot(1,3,1)
peak_times.plot(kind='bar')
plt.title('Peak Order Times')
plt.xlabel('Hour')
plt.ylabel('Orders')
#popular Dishes
plt.subplot(1,3,2)
Popular_dishes.plot(kind='bar')
plt.title("Popular Dishes")
plt.xlabel('Dish')
plt.ylabel('Orders')
#Orders by location
plt.subplot(1,3,3)
location_orders.plot(kind='pie', autopct='%1.1f%%')
plt.title('Orders by Location')
plt.ylabel('')
plt.tight_layout()
plt.show()
#for bussiness insights
print("\n-----------Bussiness Insights----------")
print("1. Peak Order Hour:",peak_times.idxmax())
print("2. Most Popular Dishes:",Popular_dishes.idxmax())
print("3. Top Location:",location_orders.idxmax())
```
